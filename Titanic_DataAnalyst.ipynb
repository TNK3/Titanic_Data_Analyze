```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Loading
df=sns.load_dataset("titanic")
print(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Step 2: Data Cleaning
# Handle Missing Values:

# Columns like age, embarked, and deck may have missing data.
print(df.isnull().sum())
```

```python
print(df.drop_duplicates())
```

```python
df.dropna(how="any")
df.dropna()
```

```python
df['sex']=df['sex']. map({'male':0,'female':1})
df.head()
```

```python
df['embarked'] = df['embarked'].map({'C': 0, 'Q': 1, 'S': 2})
df.head()
```

```python
# Survival Rate Overview:

# Calculate the overall survival rate:
survival_rate = df['survived'].mean()
print(f"Overall Survival Rate: {survival_rate * 100:.2f}%")
```

```python
# Survival by Class:

# Compare survival rates across passenger classes:
survival_by_class = df.groupby('pclass')['survived'].mean() * 100
print("Survival by class (in percentage):")
print(survival_by_class)
```

```python
# Survival by Gender:

# Analyze survival rates for males and females:
survival_by_gender = df.groupby('sex')['survived'].mean()*100
print("Survival by gender (in percentage):")
print(survival_by_gender)
```

```python
# Fare Distribution:

# Understand ticket fare distribution:
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['fare'], kde=True, bins=30)
plt.title("Fare Distribution")
plt.show()
```

```python
# Step 4: Data Visualization
# Survival Rate by Gender and Class:

sns.barplot(x="class",y="survived",hue="pclass",data=df)
plt.title("Survival rate by Gender and Class")
plt.show()
```

```python
sns.histplot(df[df['survived'] == 1]['age'], kde=True, bins=10, color='green', label='Survived')
sns.histplot(df[df['survived'] == 0]['age'], kde=True, bins=20, color='red', label='Not Survived')
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.legend()
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
